```python
%matplotlib inline
import pandas as pd
import numpy as np
import sklearn
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X, y = sklearn.datasets.make_moons(n_samples=500, noise=0.18)
```

```python
X
```

```python
y
```

```python
ds = pd.DataFrame(data=[X[:, 0], X[:, 1], y]).T
```

```python
ds
```

```python
sns.scatterplot(x=ds[0], y=ds[1], hue=ds[2])
```

```python
X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.40, random_state=42
    )
```

```python
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
```

```python
score_log = clf.score(X_test, y_test)
```

```python
score_log
```

```python
y_logreg = clf.predict(X)
```

```python
y_logreg
```

```python
sns.scatterplot(x=ds[0], y=ds[1], hue=y_logreg)
```

```python
clf = DecisionTreeClassifier(criterion='entropy', max_depth=5)
```

```python
clf.fit(X_train, y_train)
```

```python
score_tree = clf.score(X_test, y_test)
```

```python
score_tree
```

```python
y_dectree = clf.predict(X)
```

```python
y_dectree
```

```python
sns.scatterplot(x=ds[0], y=ds[1], hue=y_dectree)
```

```python
clf = RandomForestClassifier(max_depth=5, n_estimators=10)
```

```python
clf.fit(X_train, y_train)
```

```python
score_forest = clf.score(X_test, y_test)
```

```python
score_forest
```

```python
y_randforst = clf.predict(X)
```

```python
y_randforst
```

```python
sns.scatterplot(x=ds[0], y=ds[1], hue=y_randforst)
```

```python
clf = KNeighborsClassifier(3)
```

```python
clf.fit(X_train, y_train)
```

```python
score_kneighbors = clf.score(X_test, y_test)
```

```python
score_kneighbors
```

```python
y_kneighbors = clf.predict(X)
```

```python
y_kneighbors
```

```python
sns.scatterplot(x=ds[0], y=ds[1], hue=y_kneighbors)
```

```python
clf = GaussianProcessClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
score_gaussian = clf.score(X_test, y_test)
```

```python
score_gaussan
```

```python
y_gaussan = clf.predict(X)
```
